# Transformaciones Morfológicas Más Usadas

Las transformaciones morfológicas son operaciones no lineales basadas en teoría de conjuntos, 
utilizadas principalmente en imágenes binarias y en escala de grises. 
Se basan en el uso de un **elemento estructurante (kernel)** que define la vecindad de operación.

---

## 1. Erosión (Erosion)

**¿En qué consiste?**  
Reduce o adelgaza los objetos en la imagen.  
Un píxel permanece activo solo si el elemento estructurante cabe completamente dentro del objeto.

**Efecto:**
- Reduce el tamaño de objetos.
- Elimina pequeños ruidos blancos.
- Separa objetos ligeramente conectados.

**Intuición matemática (binaria):**
A ⊖ B = { z | B_z ⊆ A }

---

## 2. Dilatación (Dilation)

**¿En qué consiste?**  
Expande o engrosa los objetos.  
Un píxel se activa si al menos una parte del elemento estructurante coincide con el objeto.

**Efecto:**
- Aumenta el tamaño de objetos.
- Rellena pequeños huecos.
- Conecta regiones cercanas.

**Intuición matemática (binaria):**
A ⊕ B = { z | (B̂)_z ∩ A ≠ ∅ }

---

## 3. Apertura (Opening)

**Definición:**
Erosión seguida de dilatación con el mismo elemento estructurante.

A ○ B = (A ⊖ B) ⊕ B

**¿Para qué sirve?**
- Elimina ruido pequeño.
- Suaviza contornos.
- Mantiene el tamaño general de objetos grandes.

---

## 4. Cierre (Closing)

**Definición:**
Dilatación seguida de erosión.

A ● B = (A ⊕ B) ⊖ B

**¿Para qué sirve?**
- Rellena pequeños huecos.
- Cierra pequeñas grietas.
- Une regiones cercanas.

---

## 5. Gradiente Morfológico

**Definición:**
Diferencia entre dilatación y erosión.

Grad(A) = (A ⊕ B) − (A ⊖ B)

**¿Qué hace?**
- Resalta bordes.
- Extrae contornos de los objetos.

---

## 6. Top-Hat (White Top-Hat)

**Definición:**
Diferencia entre la imagen original y su apertura.

TopHat(A) = A − (A ○ B)

**¿Para qué sirve?**
- Detectar objetos brillantes pequeños.
- Resaltar detalles claros sobre fondo oscuro.

---

## 7. Black-Hat

**Definición:**
Diferencia entre el cierre y la imagen original.

BlackHat(A) = (A ● B) − A

**¿Para qué sirve?**
- Detectar objetos oscuros pequeños.
- Resaltar detalles oscuros sobre fondo claro.

---

# Nota Final

Todas estas operaciones dependen críticamente del:
- Tipo de imagen (binaria o escala de grises).
- Forma y tamaño del elemento estructurante.
- Objetivo del procesamiento (segmentación, limpieza, detección de bordes, etc.).

```python
# Erosion
kernel = np.ones((3,3), np.uint8)
imagen_erosion = cv2.erode(imagen_binaria, kernel, iterations=20)
cv2.imshow("Imagen Erosionada", imagen_erosion)
cv2.waitKey(0)
```

```python
# Dilatación
kernel = np.ones((5,5), np.uint8)
imagen_erosion = cv2.dilate(imagen_binaria, kernel, iterations=20)
cv2.imshow("Imagen Dilatada", imagen_erosion)
cv2.waitKey(0)
```

```python
kernel = np.ones((5,5), np.uint8)
imagen_erosion = cv2.morphologyEx(imagen_binaria, cv2.MORPH_GRADIENT, kernel)
cv2.imshow("Imagen Gradiente", imagen_erosion)
cv2.waitKey(0)
```

```python
# Kernel Elliptico
kernel_eliptico = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
kernel_eliptico
```

```python
kernel = np.ones((5,5), np.uint8)
imagen_erosion = cv2.morphologyEx(imagen_binaria, cv2.MORPH_GRADIENT, kernel_eliptico)
cv2.imshow("Imagen Gradiente", imagen_erosion)
cv2.waitKey(0)
```
