```python
import cv2
import numpy as np
```

# Operaciones Morfológicas

```python
imagen = cv2.imread("img/figuras.jpg", cv2.IMREAD_GRAYSCALE)
# Threshold convierte una imagen en escala de grises en binaria (Sólo 0's y 1's)
ret, imagen_binaria = cv2.threshold(imagen, 3*255/4, 255.0, cv2.THRESH_BINARY_INV)
cv2.imshow("Imagen", imagen_binaria)
cv2.waitKey(0)
```

```python
kernel = np.ones((5,5), np.uint8)
imagen_erosion = cv2.erode(imagen_binaria, kernel, iterations=20)
cv2.imshow("Imagen Erosionada", imagen_erosion)
cv2.waitKey(0)
```

```python
kernel = np.ones((5,5), np.uint8)
imagen_erosion = cv2.dilate(imagen_binaria, kernel, iterations=5)
cv2.imshow("Imagen Dilatada", imagen_erosion)
cv2.waitKey(0)
```

```python
kernel = np.ones((5,5), np.uint8)
imagen_erosion = cv2.morphologyEx(imagen_binaria, cv2.MORPH_GRADIENT, kernel)
cv2.imshow("Imagen Gradiente", imagen_erosion)
cv2.waitKey(0)
```
