# Bordes Canny

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_gray = cv2.imread("img/piezas.jpeg", cv2.IMREAD_GRAYSCALE)
img_edges = cv2.Canny(img_gray, 100, 200) 

plt.imshow(img_edges, cmap="gray")
plt.show()
```
