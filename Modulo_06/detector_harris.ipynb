# Detección de Características en Imágenes

```python
import cv2
import numpy as np
from skimage.feature import peak_local_max
```

```python
img = cv2.imread('img/cuadro.jpg', cv2.IMREAD_GRAYSCALE)
img = np.float32(img)

harris_corners = cv2.cornerHarris(img,2,3,0.04)
dst = cv2.dilate(harris_corners, None)

maxim = peak_local_max(dst, min_distance=1)
for coord in maxim:
    img[coord[0], coord[1]] = 255
```
