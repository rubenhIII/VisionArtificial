# Detección de Características en Imágenes
## Detectores de Esquinas

```python
import cv2
import numpy as np
from skimage.feature import peak_local_max
```

### Parameters
* src	Input single-channel 8-bit or floating-point image.
* blockSize	Neighborhood size.
* ksize	Aperture parameter for the Sobel operator.
* k	Harris detector free parameter. See the formula above.
* borderType	Pixel extrapolation method. See BorderTypes. BORDER_WRAP is not supported.

```python
# Zamora
# blockSize = 2
# ksize = 3
# k = 0.04
# threshold = 0.05

# Cuadro
# blockSize = 4
# ksize = 3
# k = 0.04
# threshold = 0.05
```

```python
blockSize = 4
ksize = 9
k = 0.04
threshold = 0.001

img = cv2.imread('img/zamora.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

harris_corners = cv2.cornerHarris(gray, blockSize, ksize, k)
harris_dilate = cv2.dilate(harris_corners, None, iterations=1)

img[harris_dilate > threshold * harris_dilate.max()]=[0,0,255]

cv2.imshow("Corners", img)
cv2.waitKey(0)
```

```python
img2 = cv2.imread('img/zamora.jpg')
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

detector = cv2.FastFeatureDetector_create(
    threshold = 80,
    nonmaxSuppression = True,
    type = cv2.FastFeatureDetector_TYPE_7_12
)
pts = detector.detect(img2_gray, mask=None)
img3 = cv2.drawKeypoints(img2, pts, None, color=(0,0,255))

cv2.imshow("Corners", img3)
cv2.waitKey(0)
```
