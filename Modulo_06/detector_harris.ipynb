# Detección de Características en Imágenes

```python
import cv2
import numpy as np
from skimage.feature import peak_local_max
```

### Parameters
* src	Input single-channel 8-bit or floating-point image.
* blockSize	Neighborhood size.
* ksize	Aperture parameter for the Sobel operator.
* k	Harris detector free parameter. See the formula above.
* borderType	Pixel extrapolation method. See BorderTypes. BORDER_WRAP is not supported.

```python
blockSize = 2
ksize = 3
k = 0.04
threshold = 0.05
```

```python
img = cv2.imread('img/chessboard.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

harris_corners = cv2.cornerHarris(gray, blockSize, ksize, k)
harris_dilate = cv2.dilate(harris_corners, None, iterations=3)

img[harris_dilate > threshold * harris_dilate.max()]=[0,0,255]
```
